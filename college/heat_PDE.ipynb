$$u_t=Du_{xx}\\s=\frac{Dk}{h^2}\\
\\w_{i,j+1}=sw_{i+1,j}+(1-2s)w_{i,j}+sw_{i-1,j}

```python
import numpy as np
import cufflinks as cf
import plotly.graph_objects as go
import pandas as pd

h=0.05#位移
k=0.001#时间
#k>0.05会使得误差爆炸
D=1
Nx=int(1/h)
Nt=int(1/k)
U=np.zeros((Nx,Nt))
U[:,0]=np.power(np.sin(2*np.pi*np.linspace(0,1,Nx)),2)
U[-1,0]=0
s=(D*k)/(h**2)
for i in range(1,Nt):
    for j in range(1,Nx-1):
        U[j,i]=s*(U[j-1,i-1]+U[j+1,i-1])+(1-2*s)*U[j,i-1]
fig=go.Figure()

X=np.linspace(0,1,Nx)
T=np.linspace(0,1,Nt)
fig.add_trace(go.Surface(x=X,y=T,z=U))
```

```python
U=np.zeros((Nx,Nt))
U[:,0]=np.power(np.sin(2*np.pi*np.linspace(0,1,Nx)),2)
U[-1,0]=0
# k=0.05
k=0.001
s=(D*k)/(h**2)
A=np.diag(1+2*s*np.ones(Nx))+np.diag(-s*np.ones(Nx-1),k=1)+np.diag(-s*np.ones(Nx-1),k=-1)
A_=np.linalg.inv(A)
B=np.zeros(Nx)
B[0],B[-1]=U[0,0],U[-1,0]
for i in range(1,Nt):
    U[:,i]=np.matmul(A_,U[:,i-1]+s*B)
U
fig=go.Figure()
X=np.linspace(0,1,Nx)
T=np.linspace(0,1,Nt)
fig.add_trace(go.Surface(x=X,y=T,z=U))
#隐式解法会收敛不会爆炸
```
