```python
import numpy as np
import plotly.graph_objects as go

z=2+5j
t=3
f=2
T=np.linspace(0,t,1000)
fig=go.Figure()
Y=np.sin(2*np.pi*f*T)
Y=Y-Y.min()#恒为正
fig.add_trace(go.Scatter(x=T,y=Y))
```

```python
f2=2
rez=Y*np.cos(-2*np.pi*f2*T)
imz=Y*np.sin(-2*np.pi*f2*T)
fig=go.Figure()
fig.add_trace(go.Scatter(x=rez,y=imz,name="图像"))
fig.update_layout(
    height=700,
    width=700)
```

$$g(t)e^{-2\pi ft\text{i}}$$

```python
def get_center(f,f2):
    x=(Y*np.cos(-2*np.pi*f2*T)).mean()#分解，三角函数的正交性
    y=(Y*np.sin(-2*np.pi*f2*T)).mean()
    return x,y
get_center(f,f2)
```

$$z_c=\frac1N\sum_{k=0}^{N}g(t_k)e^{-2\pi ft_k	\text{i}}

```python
fig.add_trace(go.Scatter(x=[get_center(f,f2)[0]],y=[get_center(f,f2)[1]],name="质心"))
```

$$\hat{g}(f)=\int_{0}^{t_m}g(t)e^{-2\pi ft	\text{i}}\text{d}t

```python
fm=5
num=500
F=np.linspace(0,fm,num)#质心的y坐标
fig=go.Figure()
CMy=[get_center(f,F[i])[1] for i in range(num)]#虚频曲线

fig.add_trace(go.Scatter(x=F,y=CMy))
```

```python
CMl=[np.sqrt(get_center(f,F[i])[1]**2+get_center(f,F[i])[0]**2) for i in range(num)]
fig=go.Figure()
fig.add_trace(go.Scatter(x=F,y=CMl))
```
