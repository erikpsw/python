```python
#拉格朗日插值
data=[[0,-1],[1,1],[2,2],[3,2],[4,6],[5,4]]
data_x=[t[0] for t in data]
data_y=[t[1] for t in data]


#计算分母denominator

#计算p(x)
def px(k,x,data_x):
    n=len(data_x)
    ans=1
    for j in range(n):
        if j==k:continue
        ans*=(x-data_x[j])
    return ans#闭包
def L(data_x,data_y,x):
    n=len(data)
    deno=[]
    for i in range(n):
        ans=1
        for j in range(n):
            if i==j:continue
            ans*=(data_x[i]-data_x[j])
        deno.append(ans)
    value=0
    for i in range(n):
        value+=data_y[i]*px(i,x,data_x)/deno[i]
    return(value)
L(data_x,data_y,4)

import numpy as np
import plotly.graph_objects as go

fig=go.Figure()
data=np.array(data)
fig.add_trace(go.Scatter(x=data[:,0],y=data[:,1],mode="markers"))
X=np.linspace(data[:,0].min(),data[:,0].max(),500)
fig.add_trace(go.Scatter(x=X,y=L(data_x,data_y,X),mode="lines"))
```

```python
diff=[data_y]
for i in range(len(data_x)-1):
    tmp=[]
    cur=diff[i]
    for j in range(len(diff[i])-1):
        tmp.append((cur[j+1]-cur[j])/(data_x[j+1+i]-data_x[j]))
    diff.append(tmp)
diff
```

```python
Y=np.zeros_like(X)
Y+=diff[0][0]
for i in range(1,len(diff)):
    tmp=diff[i][0]*(X-data_x[0])
    for j in range(1,i):
        tmp*=(X-data_x[j])
    Y+=tmp
fig.add_trace(go.Scatter(x=X,y=Y,mode="lines"))
```
