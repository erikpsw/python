```python
import numpy as np
import pandas as pd
import cufflinks as cf
from plotly import graph_objects as go

N=500
t=6*np.pi
T=np.linspace(0,t,N)
b=np.sin(T)

def my_fft(b,t):
    N=len(b)
    s = np.fft.fft(b)
    f=np.linspace(0,N/t,N)[0:int(N/2)]
    A=(2*np.sqrt(s[:].imag**2+s[:].real**2)[0:int(N/2)])/N
    return pd.Series(A,index=f)

pd.Series(b).iplot()
```

```python
my_fft(b,t).iplot()
```

```python
fig = go.Figure().set_subplots(2,2,specs=[[{}, {}],
           [{"colspan": 2}, None]],#延长
    subplot_titles=("信号1","信号2", "合成"))

N=1000
t2=np.pi
T=np.linspace(0,t2,N)
A1=2
A2=3
f1=1
f2=20#频率相差小会相互干扰
fig.add_trace(go.Scatter(x=T,y=A1*np.sin(2*np.pi*f1*T)))
fig.add_trace(go.Scatter(x=T,y=A2*np.sin(2*np.pi*f2*T)),row=1,col=2)
signal=A1*np.sin(2*np.pi*f1*T)+A2*np.sin(2*np.pi*f2*T)
fig.add_trace(go.Scatter(x=T,y=signal),row=2,col=1)

fig.update_xaxes(title_text="t/s")
fig.update_yaxes(title_text="A/m")
fig.update_layout(height=700)
```

```python
my_fft(signal,t2).iplot()
```
