```python
C=[]
def combine(n,cur,A):
    if(cur==n):
        res=[]
        for i in range(n):
            if (A[i]):
                res.append(i)
        print(res)
        if len(res)==2:
            C.append(set(res))
        return
    A[cur]=1
    combine(n,cur+1,A)
    A[cur]=0
    combine(n,cur+1,A)

combine(N,cur=0,A=[0 for i in range(N)])
```

```python
x=[0.1,0.7,0.8,0.8,0.05]

def judge(x):
    ok=0
    for i in C:
        pos=[]
        for j in i:
            pos.append(x[j])
        mi=min(pos)
        ma=max(pos)#å¼§é•¿çš„èŒƒå›´
        
        if(ma-mi<0.5):
            #case 1
            count=0
            for k in x:
                if k>=mi and k<=ma:
                    count=count+1
            if count==N:
                # print("min="+str(mi)+"max="+str(ma))
                ok=1
        elif(1+mi-ma<0.5):
            #case 2
            count=0
            for k in x:
                if k<=mi or k>=ma:
                    count=count+1
            if count==N:
                # print("min="+str(mi)+"max="+str(ma))
                ok=1
    if(ok):
        return(ok)

print(judge(x))
```

```python
yes=0
for i in range(turns):    
    if(judge(np.random.rand(5))):
        yes=yes+1
print(yes/turns)
```
