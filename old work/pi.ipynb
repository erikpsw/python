```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import matplotlib.patches as mpathes

n=500
N=0
yes=0
li=[]
Pi=[]

fig, ax = plt.subplots(figsize=(14,14))
for i in range(n):
    N=N+1
    x=random.random()*2-1
    y=random.random()*2-1
    li.append((x,y))
    plt.scatter(x,y)
    if(x**2+y**2<1):
        yes=yes+1
    Pi.append((yes/N)*4)
circle = mpathes.Circle(np.array([0,0]),1,fill=False)
ax.add_patch(circle)
rect = mpathes.Rectangle(np.array([-1,-1]),2,2,color='r',fill=False)
ax.add_patch(rect)

plt.xlabel("x") 
plt.ylabel("y") 
plt.show()
```

```python
S=pd.Series(Pi)
S
```
