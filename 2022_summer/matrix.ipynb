```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot(m,a):
    ax=plt.axes()
    ax.quiver(0,0,a[0],a[1],angles='xy', scale_units='xy', scale=1)
    ax.grid()
    ax.set_aspect(1)
    b=np.dot(m,a)
    x=max(abs(a[0]),abs(b[0]))
    y=max(abs(a[1]),abs(b[1]))
    ax.set_xlim(-x, x)
    ax.set_ylim(-y, y)
    ax.quiver(0,0,b[0], b[1],angles='xy', scale_units='xy', scale=1)
    
```

```python
a=np.array([3,2])
m=np.array([[1,-2],[-2,1]])
m
```

```python
plot(m,a)#任意线性变换
```

```python
import math
pi=math.pi
theta=pi/3
m2=np.array([[math.cos(theta),-math.sin(theta)],
[math.sin(theta),math.cos(theta)]])
plot(m2,a)#旋转矩阵
```

$\mathbf{R}_{\theta}=\left(\begin{array}{l}
\cos \theta & -\sin \theta \\
\sin \theta & \cos \theta
\end{array}\right)$

```python
lamb=0.5
m3=np.array([[1,0],[0,lamb]])
plot(m3,a)
```

```python
#反射变换

m4=np.array([[1,0],[0,-1]])
plot(m4,a)
```
