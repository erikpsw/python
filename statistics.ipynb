```python
import matplotlib.pyplot as plt
from IPython.display import Latex
import numpy as np
from fractions import Fraction
import math

fig = plt.figure(tight_layout=True)
x=np.array([-5,0,4,7,12,15,19,23,27,31,36])
y=np.array([156,150,132,128,130,116,104,89,93,76,54])

def fit(x,y):
    min=x.min()-3
    max=x.max()+3
    n=len(x)
    avg_x=sum(x)/n
    avg_y=sum(y)/n
    x_times_y=sum(x*y)
    x_2=sum(x**2)
    y_2=sum(y**2)
    avg_x_avg_y=avg_x*avg_y
    avg_x_2=avg_x**2
    avg_y_2=avg_y**2

    c=x_times_y-(n*avg_x_avg_y)
    d=x_2-n*(avg_x_2)
    # b=round(c/d)
    b=c/d
    # a=round(avg_y-(avg_x*b))
    a=avg_y-(avg_x*b)
    graph_x=np.linspace(min,max,50)
    graph_y=b*graph_x+a
    r=(x_times_y-n*avg_x*avg_y)/(math.sqrt(x_2-n*avg_x_2)*math.sqrt(y_2-n*avg_y_2))
    return{'k':b,'b':a,'x':graph_x,'y':graph_y,'r':r}
res=fit(x,y)
print("k="+str(res['k']))
print("b="+str(res['b']))
```

```python
Latex(r"$y=kx+b$")
```

```python

plt.plot(res['x'],res['y'],color='red',linewidth=1.0,linestyle='--')
plt.scatter(x, y, color='b')
a=res['b']
b=res['k']
if (a>0):
    plt.title("y="+"%.2f" %b+"x+"+"%.2f" %a)
elif(a==0):
    plt.title("y="+"%.2f" %b)
else:
    plt.title("y="+"%.2f" %b+"x"+"%.2f" %a)
plt.show()
```

```python
y0=np.array([220,223,222,261,273,309,394,475,525,620,719,849,917,1210,1487,1510,1827,1837,1949,2492,2661])
x=np.array(range(1,len(y0)+1))
y=np.log(y0)
res=fit(x,y)
print(res)
```

```python

b=res['k']
print("b="+str(res['k']))
a=math.e**res['b']
print("a="+str(math.e**res['b']))
```

```python
plt.plot(res['x'],a*(np.exp(b*res['x'])),color='red',linewidth=1.0,linestyle='--')
plt.scatter(x, y0, color='b')
a="%.2f" %math.e**(res['b'])
b="%.2f" %res['k']
plt.title("$y="+str(a)+"e^{"+str(b)+"x}$")
plt.show()
```
